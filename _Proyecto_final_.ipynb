# b) Variables categóricas y balanceo


```python
import matplotlib.pyplot as plt

categorical_cols = df.select_dtypes(include=['object', 'category']).columns
print(categorical_cols)

for col in categorical_cols:
    fig, ax = plt.subplots(figsize=(5, 5))

    df[col].value_counts(dropna=False).plot.pie(
        autopct='%1.1f%%',
        ax=ax,
        shadow=False,
        textprops={'fontsize': 12}
    )
    ax.set_title(f'Gráfico de pastel para la variable "{col}"', fontsize=12)
    ax.set_ylabel('')

    plt.tight_layout()
    plt.show()
```

```python
# Implicaciones del balanceo o desbalanceo en las variables:
# si una varibal esta balanceada, podemos asegurar que  las proximas funciones como ANOVA van a comparar de manera justo cada una de las variable, los resultados van a tener mayor validez y se puede evitar el sesgo
# si una variable no esta balanceada, ANOVA pierde poder estadistico, no hay sufiicente informacion para comparar y predominan los sesgos
```

# **c) Modelo ANOVA**

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
# Y ~ X
modelo_lineal = ols('precio ~ costo + ventas + margen_beneficio + inventario + porcentaje_descuento + dias_entrega + C(categoria) + C(material) + C(color) + C(ubicacion) + C(temporada) + C(tipo_tienda) + C(marca) + ingresos', data = df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
```

```python
# Se comparan los diferentes grupos en relacion al precio para determinar cual influye mas en en este ultimo, tambien nos permite descartar variables con poco aporte
# Costo y Margen de Beneficio son las mas relevantes
```

# **d) Test de Tukey y codificación**

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(
    endog=df['precio'], # Variable de respuesta
    groups=df['categoria'],          # Variable categorica
    alpha=nivel_de_significancia) # nivel de significancia
print(tukey)

tukey.plot_simultaneous(ylabel="categoria", xlabel="precio")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

# **f) Modelo matemático de precio de venta**


```python
x = df[["categoria","material","color","ubicacion","temporada", "marca", "ventas", "margen_beneficio", "ingresos", "costo"]]
y = df["precio"]


df_cleaned = df[x.columns.tolist() + [y.name]].dropna()
x_cleaned = df_cleaned[x.columns]
y_cleaned = df_cleaned[y.name]

#Recta de regresión ajustada
import statsmodels.api as sm
x_constante = sm.add_constant(x_cleaned)
modelo = sm.OLS(y_cleaned, x_constante).fit()

yc = modelo.predict(x_constante)

#Coeficiente de determinación
from sklearn.metrics import r2_score  # recomendada
r2 = r2_score(y_cleaned, yc)
print(f'Coeficiente de determinación: {r2: 0.4f}\n')
```

```python
modelo.params
```

Ecuación para sacar precio

-104.791595+1.385822 (costo) - 0.080882 (ventas) + 3.844568(margen_beneficio) - 0.520113 (categoria) + 0.164294 (material) - 0.094296 (color) - 0.281479 (ubicacion) - 0.495307 (temporada) - 0.520500 (marca) + 0.000321 (Ingresos)
